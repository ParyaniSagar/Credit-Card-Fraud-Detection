```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit_card_data = pd.read_csv('../2_Dataset/creditcard.csv')
```

```python
credit_card_data.head()
```

```python
credit_card_data.info()
```

```python
credit_card_data.describe()
```

```python
credit_card_data.isnull().sum()
```

```python
credit_card_data['Class'].value_counts()
```

```python
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# to balance out unbalanced data
legit_sample = legit.sample(n=492)
```

```python
new_data = pd.concat([fraud,legit_sample],axis=0)
new_data.head()
```

```python
new_data.groupby('Class').mean()
```

```python
features = new_data.drop(columns='Class',axis=1)
labels = new_data['Class']
```

```python
features_train,features_test,labels_train,labels_test = train_test_split(features,labels,test_size=0.2,stratify=labels,random_state=2)
```

```python
# Model training
model = LogisticRegression()
model.fit(features_train,labels_train)
```

```python
features_train_pred = model.predict(features_train)
training_accuracy = accuracy_score(features_train_pred,labels_train)
```

```python
print(training_accuracy)
```

```python
features_test_pred = model.predict(features_test)
test_accuracy = accuracy_score(features_test_pred,labels_test)
```

```python
print(test_accuracy)
```
